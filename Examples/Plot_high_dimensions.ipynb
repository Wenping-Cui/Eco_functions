## Explore Community augmentation, Rejection failure, Indirect failure, Replacement
* Community augmentation: the invasion is successful and the community’s richness is increased by one.
* Rejection failure: the invasive specie failed and the resident community stay the same.
* Indirect failure: the invasive specie failed but the species or resource abundance is changed.
* Replacement: the invasion is successful and one or more resident species extinct.
